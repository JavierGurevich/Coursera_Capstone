```python
#First, I download the requires packages
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Assign the url to a variable
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
#Request the content of the URL
res=requests.get(url).content
```

```python
#Create the BeautifulSoup object
soup=BeautifulSoup(res,'html.parser')
```

```python
#Create the table by searching the table in the wikipedia page by it's class
tables=soup.find_all("table",{"class":["sortable","plainrowheaders"]})
```

```python
#Make empty lists to fill later on with the values of the table
Postcode=[]
Borough=[]
Neighborhood=[]
Table=[]
```

```python
#Fill the empty lists with the values of the table's cells
for table in range(len(tables)):
    x=tables[table].find("tr").find_next("tr")
    y=x.find_all("td")
    Table.append(len(y))
    
    for row in tables[table].find_all("tr"):
        try:
            Postcode.append(row.find("td").text)
            Borough.append(row.find("td").find_next("td").text)
            Neighborhood.append(row.find("td").find_next("td").find_next("td").text)
        except:
            pass
```
